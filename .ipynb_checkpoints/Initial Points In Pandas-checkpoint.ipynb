# Test Data Generation
Script is to provide some test data for the trialateration algorithm.

Intial values are 500,500, -50

```python
import pandas as pd
import numpy as np

#create sample points
dataX = pd.DataFrame(np.random.randint(0,1000,size=(1000,1)), columns=list('X'))
dataY = pd.DataFrame(np.random.randint(0,1000,size=(1000,1)), columns=list('Y'))
dataZ = pd.DataFrame(np.random.randint(-100,0,size=(1000,1)), columns=list('Z'))

data = pd.concat([dataX, dataY, dataZ], axis=1, sort=False)


#intial values
Xi=500
Yi=500
Zi=-100


#distances
data['dist'] = ((data['X']-Xi)**2 + (data['Y']-Yi)**2 + (data['Z']-Zi)**2)**0.5



```
